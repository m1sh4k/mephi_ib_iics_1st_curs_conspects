```python
from math import log
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi, 100)  # x values from 1 to 16
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
#ax.set_xscale('log', base=2)  # set x-axis to log base 2
ax.axis('off')  # hide axes
plt.savefig('sin_x.svg')
plt.show()
```
